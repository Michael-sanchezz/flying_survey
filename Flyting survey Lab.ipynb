```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fly_df = pd.read_csv("C:/Users/micha/PycharmProjects/Docs/flying-etiquette.csv",delimiter=",")
display(fly_df)
```

```python
#pt1 1.Remove any blanks, duplication, 

df = fly_df.copy()
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
display(df)
```

```python
#pt1 2.Change “Under 5 ft, to 5 ft”
# df['How tall are you?'] = df['How tall are you?'].replace('Under 5 ft', "5'0\"")
df['How tall are you?'] = df['How tall are you?'].replace('Under 5 ft.', "5'0\"")

display(df)
```

```python
#pt1 3.Create a new column “Height (Inches)” that converts the “How tall are you?” column in feet and inches to an integer total of inches.
# Function to convert feet and inches to total inches

# df = pd.DataFrame(fly_df)

# Function to convert height from feet and inches to inches
def convert_to_inches(height_str):   
    feet,inches = height_str.split("'")
    inches = inches.replace('"', "")  
    total_inches = feet * 12 + inches
    return total_inches

# Apply the function to create a new column "Height (Inches)"
df["Height (Inches)"] = df["How tall are you?"].apply(convert_to_inches)
display(df)
```

```python
# pt1 4.Check for outliers. 
from scipy import stats


# Check for outliers in numeric columns
numeric_cols = df.select_dtypes(include='number').columns
outliers = {}
for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Visualize the data
plt.figure(figsize=(10, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, len(numeric_cols), i)
    plt.boxplot(df[col], vert=False)
    plt.title(col)
plt.suptitle("Boxplot of Flying Etiquette Survey Responses")
plt.tight_layout()
plt.show()

# Print outliers
for col, outlier_df in outliers.items():
    if not outlier_df.empty:
        print(f"\nOutliers in column '{col}':")
        print(outlier_df)
display(df)
```

```python
# pt2 1.Percent breakdown of survey participants by age range.
age_ranges = ['18-29', '30-44', '45-60', '> 60']
total_participants = df['Age'].count()

age_counts = df['Age'].value_counts().reindex(age_ranges, fill_value=0)
percentages = (age_counts / total_participants) * 100

# Plot the data
plt.figure(figsize=(8, 6))
plt.bar(percentages.index, percentages.values, color='skyblue')
plt.title('Percent Breakdown of Survey Participants by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Percentage')
plt.ylim(0, 40)  # Set y-axis limit for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7) #Adds gridlines to the plot for the y-axis.
plt.xticks(rotation=45) #Rotates the x-axis labels by 45 degrees for better readability.
for i, v in enumerate(percentages.values):
    plt.text(i, v , f"{v:.1f}%", ha='center')# Adds text labels showing the exact percentage value for each bar.
    # ha='center': Specifies the horizontal alignment of the text label. In this case,
    # i is the index of the age range (0, 1, 2, ...) within the percentages array
     # v is the percentage value corresponding to the height of the bar for the current age range.
plt.tight_layout()#Adjusts the layout of the plot to ensure everything fits properly.
plt.show()
```

```python
#pt2 2. How did survey respondents generally feel regarding if “it [was] rude to ask someone to switch seats 
# with you in order to be closer to friends?”?

responses = df['ask to move closer to friends rude?'].value_counts(normalize=True) * 100

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(responses.index, responses.values, color='green')
# responses.index is the index of the responses Series, which contains the different response options 
# responses.values: This represents the y-values of the bar chart. It contains the corresponding percentages of respondents for each response option.

plt.title("Survey Respondents' Opinions on Asking Someone to Switch Seats to Be Closer to Friends")
plt.xlabel("Opinion")
plt.ylabel("Percentage")
plt.ylim(0, 90)  # Setting y-axis limit
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding gridlines for better readability

# Adding percentage labels on top of bars
for i, v in enumerate(responses.values):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center', va='bottom')

plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# pt2 3.Use the respondents age-range ( treat as a category) as a dimension to see if there are any differences to responses to
# “it [was] rude to ask someone to switch seats with you in order to be closer to friends?”?
# Create a crosstabulation (contingency table) between Age and Seatmate
cross_table = pd.crosstab(df['Age'], df['ask to move closer to friends rude?'], normalize='index') * 100

# Plotting
plt.figure(figsize=(10, 6))

# Plotting each age range separately
for age_range in cross_table.index:
    plt.bar(cross_table.columns, cross_table.loc[age_range], label=age_range)
for i, v in enumerate(responses.values):
    plt.text(i%3, v + 1, f"{v:.1f}%", ha='center', va='bottom')

# Adding labels and title
plt.title("Responses to 'Is it rude to ask someone to switch seats to be closer to friends?' by Age Range")
plt.xlabel("Response")
plt.ylabel("Percentage")
plt.legend(title="Age Range")
plt.ylim(0,90)
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
